```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
dataset = pd.read_csv("PrePlacementData.csv")
dataset
```

```python
# No of rows and coumns
print(f"The number of rows = {dataset.shape[0]}")     # Index 0 is row access
print(f"The number of columns = {dataset.shape[1]}")  # index 1 is column access
```

# Univariate Analysis

```python
# Distplot
sns.distplot(dataset["ssc_p"],kde_kws={"color":'red'},hist=True,kde=True)
```

***This histogram that shows the frequency distribution of data for the 10th grade marks of the students***

# Bivariate Analysis

```python
# Jointplot
sns.jointplot(x="ssc_p",y="salary",data=dataset)
plt.show()
```

***This jointplot shows the relation between 10th grade marks and salary***

***It shows that most students get a salary between 2-4 lakhs***

***Some students get salary of 0 which represents non places students and also there are a very few students not placed when they scored above 70 percentage***

```python
sns.jointplot(x="ssc_p",y="salary",data=dataset,kind="hex")
plt.show()
```

***This is the same jointplot represented using hexagon where color shades represents the frequency of that data point***

***This shows most students score percentages between 60-75 and their salaries are around 3 lakhs***

```python
sns.jointplot(x="ssc_p",y="salary",data=dataset,kind="kde")
plt.show()
```

***This graph is represented using Kernel Desity estimation***
***There are 2 dense area in this graph where one shows the salary of 0 for non placed students and the other for students scored 60-90 percentage have got salary between 2-4 lakhs***

```python
# Pairplot
sns.pairplot(dataset,hue="gender",diag_kind="kde",kind="scatter",palette="husl")
plt.show()
```

***This pairplot shows the relationship between all the  numerical columns in the dataset***

***The diagonal which is relation between same columns is represented using KDE and other columns relation are represented using scatter plot***

# Categorical data

```python
# strip plot
sns.stripplot(x="ssc_b",y="ssc_p",data=dataset)
plt.show()
```

***This strip plot shows the relation between 10th grade marks of central and other board students***

***Both the board plots look very similar where most marks range from 50-90 percentage***

```python
sns.stripplot(x="gender",y="ssc_p",data=dataset,jitter=True)
plt.show()
```

***Jitter is  used hear to seperate all the overlapping datapoints for better visibility***

```python
sns.swarmplot(x="gender",y="ssc_p",data=dataset)
```

***This is a swarmplot which represents the distributionof  datapoints without overlapping. This is also an advanced version of strip plot using jitter***

# Distribution Observation

```python
# Vilin plot
sns.violinplot(x="gender",y="ssc_p",data=dataset,hue="gender")
```

***Violin plot shows the sistribution and density of datapoints with a combination of box plot.***

***This plot shows that female students score higher marks than male students. which is indicated by a high median(white line).***

```python
sns.violinplot(x = "specialisation", y = "salary", data = dataset,hue="gender")
plt.show()
```

***Violin plot for specialisation and salary with respect to gender***

***It represets that Marketing and finance specialisation gets higher salary than marketing and hr.***

***In both marketing and finance adn  marketing and HR stream male students get higher salary compared to females.***

```python
# Box Plot
sns.boxplot(x = "ssc_b", y = "ssc_p", data = dataset,hue="gender")
```

***This box plot shows that female students score more marks than male students in both the others and central board.***
***This also shows that other board students have score better marks than central board***

```python
# Bar plot
sns.barplot(x = "ssc_b", y = "ssc_p", data = dataset,hue="gender")
```

***This barplot shows the difference in marks between central and other board students and also difference in marks between male and female students***

```python
# Count PLot 
sns.countplot(x = "ssc_b", data = dataset,hue="gender")
```

***This is a count plot thats shows the count of male and female students in both the central and other boards***

```python
# Point Plot
sns.pointplot(x = "ssc_b", y = "ssc_p", data = dataset,hue="gender")
```

***This point plot shows that female students score higher on average than male students in both boards, though average scores for both genders drop from the Others board to the Central board.***

```python
# Box Plot
sns.boxplot(data = dataset, orient = "h")
```
