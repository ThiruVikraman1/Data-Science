# 8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.(Make decision using Hypothesis Testing)

# Hypothesis
***H0: There is no significant difference between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary***

***H1: There is significant difference between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary***

```python
from scipy.stats import ttest_ind

sci_tech_salary = dataset[dataset['degree_t'] == 'Sci&Tech']['salary']
mkt_hr_salary = dataset[dataset['specialisation'] == 'Mkt&HR']['salary']

ttest_ind(sci_tech_salary,mkt_hr_salary)
```

# Interpretation
***I have solved this using Unpaired Ttest***

***pvalue<0.05 so we reject null hypothesis.***

***Therefore there is significant difference(No similarity) between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary***

# 11)Test the similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis Testing)

# Hypothesis
***H0: There is no significant difference (similiarity) between test scores of etest_p and mba_p of degree(Sci&tech)***

***H1: There is significant difference (no similiarity) between test scores of etest_p and mba_p of degree(Sci&tech)***

```python

from scipy.stats import ttest_rel

etest = dataset[dataset["degree_t"]=="Sci&Tech"]["etest_p"]
Mba = dataset[dataset["degree_t"]=="Sci&Tech"]["mba_p"]

ttest_rel(etest,Mba)
```

# Interpretation
***I have used paired T-test for this question***

***pvalue<0.05 so we reject null hypothesis***

***Therefore, there is significant difference (no similiarity) between test scores of etest_p and mba_p of degree(Sci&tech)***
